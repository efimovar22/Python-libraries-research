````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

```plot``` - линия \
```scatter``` - точки \
```bar``` - столбцы

````python
x = np.linspace(0, 1, num=100, endpoint=True)
y = np.sin(x / 10) + np.cos(x ** 2)
a = np.random.random(100)
a[50] = 1000 # => массив [0, 1000] -> [0, 1] => 1000 -> 1, a[i] -> ~ 0.001
plt.scatter(x, y, c=a, cmap='magma')
plt.scatter(x[40], y[40], color='r') # отметим точку на графике
plt.show() # Отдельно выводим график
````

````python
x_bar = np.linspace(-20, 20, 200, endpoint=True)
y_bar = x_bar ** 2
plt.xlabel('x')
plt.ylabel('y')
plt.title('bars', fontsize=15, fontweight='bold')
plt.xticks(np.arange(-20, 21, 10))
plt.yticks(np.arange(-40, 401, 80))
plt.ylim(-40, 400)
plt.bar(x_bar, y_bar, label='y = x ^ 2')
plt.legend(loc=10)
plt.show()
````

````python
names = ['a', 'b', 'c']
values = [11, 12, 8]

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1) # графики будут расположены в таблице 1 на 3, 3е число - номер
plt.bar(names, values)
plt.grid(axis='x')

plt.subplot(1, 3, 2)
plt.plot(names, values)
plt.grid(axis='y')

plt.subplot(1, 3, 3)
plt.scatter(names, values)
plt.grid()

plt.show()
````

```hist```- гистограммы \
```boxplot``` - ящик с усами \
```violinplot``` - диаграмма-виолончель

````python
s = np.random.normal(0, 1, 1000)

plt.hist(s, bins=10, width=0.4)
plt.show()
````

````python
plt.subplot(1, 2, 1)
plt.boxplot(s)

plt.subplot(1, 2, 2)
plt.violinplot(s)

plt.show()
# центр - медиана = 0
# коробка - квартили 25, 75
# линии - перцинтили (5 - 95)
# точки - выбросы
````

## Параметры
### plot
linewidth - толщина линии \
linestyle - форма линии
### scatter
s - размер точек (можно s = x - чем больше х, чем больше точки) \
marker - что ставим (точки, треугольники, квадарты, ...)
### bar
width - толщина
### common
alpha - прозрачность

Разбиение плоскости для графиков на части: \
plt.subplot(2, 2, 1) \
plt.subplot(2, 2, 3) \
plt.subplot(1, 2, 2)
