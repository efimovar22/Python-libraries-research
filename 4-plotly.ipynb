```python
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots

import numpy as np
import pandas as pd
```

Быстро строим график с помощью px:

```python
x = np.linspace(0, 4, 100)
def f(x):
    return np.sqrt(x)

px.scatter(x=x, y=f(x)).show()
```

```python
go.Scatter(x=x, y=f(x))
```

```python
fig = go.Figure() # cоздали фигуру

fig.add_trace(go.Scatter(x=x, y=f(x), mode='lines+markers', name='f(x)=x<sup>0.5</sup>', marker=dict(color='orange')))

fig.add_trace(go.Scatter(x=x, y=x ** 2, name='f(x)=x<sup>2</sup>', marker=dict(color='LightSkyBlue')))

fig.update_layout(title="Plot",
                xaxis_title="x",
                yaxis_title="y",
                # legend_orientation='h',
                margin=dict(l=0, r=0, t=30, b=0),
                legend=dict(x=.5, xanchor="center"),
                hovermode='x')

fig.update_traces(hoverinfo="all", hovertemplate="Аргумент: %{x}<br>Функция: %{y}") # когда наводим курсор

fig.update_yaxes(range=[-0.2, 4], zeroline=True, zerolinewidth=2, zerolinecolor='LightPink')

fig.update_xaxes(range=[-0.2, 2], zeroline=True, zerolinewidth=2, zerolinecolor='LightPink')

fig.show()
```

```python
fig = make_subplots(rows=1, cols=2, column_widths=[1, 2]) # соотношение

fig.update_yaxes(range=[-0.5, 16], zeroline=True, zerolinewidth=2, zerolinecolor='LightPink', col=2) # row = 1 default

fig.update_xaxes(range=[-0.5, 4], zeroline=True, zerolinewidth=2, zerolinecolor='#008000', col=2)

fig.add_trace(go.Scatter(x=x, y=x,  name='h(x)=x'), 1, 1)

fig.add_trace(go.Scatter(x=x, y=x ** 2,  name='k(x)=x<sup>2</sup>'), 1, 1)

fig.add_trace(go.Scatter(visible='legendonly', x=x, y=1/(x + 1),  name='m(x)=1/(x + 1)'), 1, 1) # visible - показываем сначала график или нет

fig.add_trace(go.Scatter(x=x, y=x, mode='lines+markers', name='f(x)=x'), 1, 2)

fig.add_trace(go.Scatter(x=x, y=x ** 2, mode='lines+markers',  name='f(x)=x<sup>2</sup>', 
                         marker=dict(color=x ** 2, colorbar=dict(title="x<sup>2</sup>"), colorscale='Inferno', size=50*abs(x))
                        ), 1, 2)
fig.update_layout(legend_orientation="h",
                  legend=dict(x=.5, xanchor="center"),
                  hovermode="x",
                  margin=dict(l=0, r=0, t=0, b=0))
fig.update_traces(hoverinfo="all", hovertemplate="Аргумент: %{x}<br>Функция: %{y}")

fig.show()
```

```python
random_list = np.random.randint(0, 4, size=10000)
data = {
    'Количество нулей': np.count_nonzero(random_list == 0),
    'Количество единиц': np.count_nonzero(random_list == 1), 
    'Количество двоек': np.count_nonzero(random_list == 2),
    'Количество троек': np.count_nonzero(random_list == 3), 
}
df = pd.DataFrame(data, index=[0])

fig = go.Figure(data=[go.Pie(labels=df.columns, values=df.iloc[0])])
fig.show()
```
